```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales.csv")
```

```python
df=pd.read_csv("supermarket_sales.csv")
df.info()
```

```python
#get the shape of the dataframe(rows,columns)
df.shape
```

```python
df.head()
```

```python
#get summary statistics(mean,std,min,max,etc.)
df.describe()
```

```python
df.info()
```

```python
#check for missing values
df.isnull()
```

```python
#total missing values
df.isnull().sum()
```

```python
#drop rows with missing values(if appropriate)
df.dropna(inplace=True)
```

```python
df.head()
#to check if the values have actually been dropped or not(to check the output)
```

```python
#fill missing values with mean, median or other methods in places where there's null values(before dropping rows with null values)
#for now it won't run
df.fillna(df["Payment"].mean(),inplace=True)
```

```python
#histograms-visualize the distribution of numerical values
sns.histplot(df["Quantity"])
plt.show()
```

```python
#box plots- identify outliners and understand the spread of data
sns.boxplot(x='Payment', y='Unit price', data=df)
plt.show()
```

```python
#scatter plots- examine relationships between two numerical variables
sns.scatterplot(x='cogs', y='Unit price', data=df)
plt.show()
```

```python
#bar charts- visualize the distribution of categorial data
sns.countplot(x='Payment', data=df)
plt.show()
```

```python
#correlatiopn matrix- understand the relationships between multiple numerical variables
af=df[['Unit price', 'Quantity', 'Tax 5%', 'cogs', 'gross income', 'Rating']]
corr = af.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#groupby- analyse data based on different categories
df.groupby('City')['Quantity'].mean()
```

```python
#create a pivot table
pd.pivot_table(df, values='Quantity', index='City', columns='Payment')
```
